Python machine learning module scikit-learn will support the option to easily create learning curves in version 0.15. Learning curves allow you to see if your model will get better if you had more training data. Over the course of a weekend I played around with this new feature and wrote up this short post.

Now we can get down to business. We'll use the digits dataset.

```python
from sklearn.datasets import load_digits
digits=load_digits()
print "Dimensions of digit images: %s" % (digits.images.shape,)
```

Let's see what these digit images look like.

```python
import matplotlib.pyplot as plt
# we need this to display images inline
%pylab inline
_, ax = plt.subplots(1, 10, figsize=(15,2))
for i in range(0,10):
    ax[i].matshow(digits.images[i,:,:], cmap=plt.cm.Greys)
_, ax = plt.subplots(1, 10, figsize=(15,2))
for i in range(0,10):
    ax[i].matshow(digits.images[i+10,:,:], cmap=plt.cm.Greys)
```

Normalize the features and create feature vectors:

```python
X = (digits.data / 16).reshape(-1, 64)
y = digits.target
print X.shape
print y.shape 
```

Let's try a simple model and see how it performs

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
print X_train.shape
print X_test.shape

nb = GaussianNB()
nb.fit(X_train, y_train);
print "Train accuracy: %f" % (nb.score(X_train, y_train))
print "Test accuracy: %f"% (nb.score(X_test, y_test))
```

Now let's plot a learning curve. I will make use of the `sklearn.learning_curve` function. Scores are averaged over a 5 -fold cross validation.

```python
from sklearn.learning_curve import learning_curve
plt.xlabel("# Training sample")
plt.ylabel("Accuracy")
train_sample_size, train_scores, test_scores = learning_curve(nb, 
    X_train, y_train, cv=5, n_jobs=3, train_sizes=np.arange(60, 1000, 20))
mean_train_scores = np.mean(train_scores, axis=1)
mean_test_scores = np.mean(test_scores, axis=1)
plt.plot(train_sample_size, mean_train_scores, label="Training accuracy", color="b")
plt.plot(train_sample_size, mean_test_scores, label="cross validation accuracy", color="g")
plt.legend(loc="lower right")
```

Let's pimp this a bit by including a measure of the variance of the accuracy in the graph

This looks like a classic example of a learning curve. Let's try a few other models and see how they compare. First define a convience function for drawing the plots.

```python
def plot_lurning_curve(ax, clsfr, clsfr_name, sizes):
    ax.set_title("Learning Curve ("+clsfr_name+")")
    ax.set_xlabel("# Training sample")
    ax.set_ylabel("Accuracy")
    ax.set_ylim([0, 1])
    ax.set_yticks(np.arange(0,1,0.1))
    ax.grid()
    
    train_sample_size, train_scores, test_scores = learning_curve(
        clsfr, X_train, y_train, cv=5, n_jobs=3, train_sizes=sizes)
    mean_train_scores = np.mean(train_scores, axis=1)
    mean_test_scores = np.mean(test_scores, axis=1)
    std_train_scores = np.std(train_scores, axis=1)
    std_test_scores = np.std(test_scores, axis=1)
    ax.plot(train_sample_size, mean_train_scores, label="Training accuracy", color="b")
    ax.fill_between(train_sample_size, mean_train_scores - std_train_scores,
                     mean_train_scores + std_train_scores, alpha=0.1, color="b")
    ax.plot(train_sample_size, mean_test_scores, label="Cross-validation accuracy",
             color="g")
    ax.fill_between(train_sample_size, mean_test_scores - std_test_scores,
                     mean_test_scores + std_test_scores, alpha=0.1, color="g")
    ax.legend(loc="lower right")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC

sizes = np.arange(10, 1000, 10)

rf = RandomForestClassifier()
gnb = GaussianNB()
svm = LinearSVC()
svc = SVC()

_, ax = plt.subplots(2, 2, figsize=(15, 10))

plot_lurning_curve(ax[0,0], gnb, "GaussianNB", sizes)
plot_lurning_curve(ax[0,1], svm, "LinearSVC", sizes)
plot_lurning_curve(ax[1,0], svc, "SVC", sizes)
plot_lurning_curve(ax[1,1], rf, "RandomForest", sizes)
```

For some bonus points let's optimize one of the models.

```python
from sklearn.grid_search import GridSearchCV
param_grid = {'n_estimators': np.arange(5, 50)}
grid_search = GridSearchCV(rf, param_grid=param_grid, cv=3)
grid_search.fit(X_train, y_train);
plt.figure(figsize=(15, 6))
plt.plot([c.mean_validation_score for c in grid_search.grid_scores_], label="validation score")
plt.xticks(np.arange(len(grid_search.grid_scores_)), param_grid['n_estimators']); 
plt.xlabel("Number of trees"); plt.ylabel("Accuracy");plt.legend(loc='best');
plt.grid()
```

There doesn't seem to be much to gain by using more than 30 estimators. Lets compare learning curves for different model complexities.

```python
fix, ax = plt.subplots(1, 4, figsize=(15, 6))
plot_lurning_curve(ax[0], RandomForestClassifier(n_estimators=5), "n=5", sizes)
plot_lurning_curve(ax[1], RandomForestClassifier(n_estimators=10), "n=10", sizes)
plot_lurning_curve(ax[2], RandomForestClassifier(n_estimators=30), "n=30", sizes)
plot_lurning_curve(ax[3], RandomForestClassifier(n_estimators=50), "n=50", sizes)
```
